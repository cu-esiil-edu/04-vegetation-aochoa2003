```python
# Compute the difference in NDVI before and after
ndvi_diff =(
    ndvi_da
        .sel(date=slice('2012', '2022'))
        .mean('date')
        .NDVI
    - ndvi_da
        .sel(date=slice('2001', '2012'))
        .mean('date')
        .NDVI
)
print(ndvi_diff.head())

ndvi_diff.hvplot(
    x='x',
    y='y',
    cmap='PiYG',
    geo=True,
    title='NDVI difference: 2012-2022 minus 2001-2012'
)
```
